```python
dates=[dictionary['date'] for dictionary in datalist ]
dates.sort()
dates
```

```python
def parse_date(datestring):
    """ Convert a date string into a pandas datetime object """
    return pd.to_datetime(datestring, format="%Y-%m-%d")
```

```python
startdate=parse_date(dates[0])
enddate=parse_date(dates[-1])
print (startdate, ' to ', enddate)
```

```python
index=pd.date_range(startdate, enddate, freq='D')
casebytestdf=pd.DataFrame(index=index, columns=['newCasesByPublishDate', 'cumVirusTestsByPublishDate'])
casebytestdf
```

```python
for entry in datalist:
    date=parse_date(entry['date'])
    for column in ['newCasesByPublishDate', 'cumVirusTestsByPublishDate']:
        if pd.isna(casebytestdf.loc[date, column]): 
            value= float(entry[column]) if entry[column]!=None else 0.0
            casebytestdf.loc[date, column]=value
casebytestdf.fillna(0.0, inplace=True)
            
casebytestdf
        
```

```python
def access_api(button):
    apibutton.icon="check"
    apibutton.disabled=True
apibutton=wdg.Button(
    description='data',
    disabled=False,
    button_style='',
    tooltip='Click to download current Public Health England data',
    icon='download'
)

apibutton.on_click(access_api)
display(apibutton)
```

```python
casebytestdf=pd.read_pickle("casebytestdf.pkl")
```

```python
agecols=wdg.SelectMultiple(
    options=['newCasesByPublishDate', 'cumVirusTestsByPublishDate'], 
    value=['newCasesByPublishDate', 'cumVirusTestsByPublishDate'],
    rows=2,
    disabled=False
)

def graph(graphcolumns):
    ncols=len(graphcolumns)
    if ncols>0:
        casebytestdf.plot(kind='line', y=list(graphcolumns)) 
        plt.show() 
    else:
        print("Click to select data for graph")
        print("(CTRL-Click to select more than one category)")
        
output=wdg.interactive_output(graph, {'graphcolumns': agecols})

display(agecols, output)
```

```python
series=wdg.SelectMultiple(
    options=['newCasesByPublishDate', 'cumVirusTestsByPublishDate'], 
    value=['newCasesByPublishDate', 'cumVirusTestsByPublishDate'],
    rows=2,
    disabled=False
)   
    
scale=wdg.RadioButtons(
    options=['linear', 'log'],

    description='Scale:',
    disabled=False
)
controls=wdg.HBox([series, scale])

def graph(gcols, gscale):
    if gscale=='linear':
        logscale=False
    else:
        logscale=True
    ncols=len(gcols)
    if ncols>0:
        casebytestdf[list(gcols)].plot(logy=logscale)
        plt.show() 
    else:
        print("Click to select data for graph")
        print("(CTRL-Click to select more than one category)")

    

graph=wdg.interactive_output(graph, {'gcols':series, 'gscale': scale})

display(controls, graph)
```

```python
ctrls=wdg.VBox([series, scale])
form=wdg.HBox([graph, ctrls])


display(form)
```
